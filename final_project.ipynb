```python
%matplotlib inline
#import gzip
import urllib
import urllib2
import re
import matplotlib.pyplot as plt
import numpy as np
import sympy as sp
import pandas as pd

html = urllib2.urlopen("https://datasets.imdbws.com/") # open url
t = html.read() # read html
data_links = re.findall('(https://.*\..*.tsv.gz)>', t) # find all data links
print data_links
gz_datasets = [re.search('([^/]+.tsv.gz)', i).group() for i in data_links] # extract all gz extension filenames from the data links list
print gz_datasets
n = 0 # which dataset do download
urllib.urlretrieve(data_links[n], gz_datasets[n]) # download the files
```

```python
dataset = [0 for i in data_links]
dataset[n] = pd.read_table(gz_datasets[n],'\t', compression = "gzip", nrows = 200) # file, separator
print dataset[n]
```
